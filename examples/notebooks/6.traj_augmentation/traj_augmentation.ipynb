# Augmenting Molecular Dynamics Trajectory with interpolation

### Introduction

Some trajectories might contain sparse sampling of important transitions between states. In this notebook, we will explore how we can interpolate between frames in a molecular dynamics trajectory to create a more continuous representation of the molecular motion.

```python
from deep_cartograph.tools import traj_augmentation
from deep_cartograph import data

import importlib.resources as resources

# Get the path to the data
data_folder = resources.files(data)

# Paths to trajectory and topology files
traj_path = f"{data_folder}/calpha_transitions/input/6IRS_7DSQ/GOdMD_6IRS_7DSQ.dcd"
top_path = f"{data_folder}/calpha_transitions/input/6IRS_7DSQ/GOdMD_6IRS_7DSQ.pdb"

new_traj_path, new_top_path = traj_augmentation(
    configuration={},
    trajectories=traj_path,
    topologies=top_path,
    output_folder="traj_augmentation_output",
)
```
